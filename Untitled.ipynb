```python
# Importing the required packages.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings

# Ignoring warnings for a cleaner output.
warnings.filterwarnings('ignore')

# Setting the visual style for the plots.
sns.set()

# Importing the data and getting rid of the rank column.
# This is game by game data for the 2023-2024 full season.
data = pd.read_csv('https://github.com/noah-po/Personal-Stuff/raw/main/Sports%20Stuff.csv')
data = data.iloc[:,1:]

# Function to get player data for a given player ID.
def get_player_data(player_id, overlapped_dates):
    player_data = overlapped_dates[overlapped_dates['Player-additional'] == player_id].reset_index(drop = True).reset_index()
    return player_data

# Function to calculate and plot the correlation matrix between two players.
def playerCorrelations():
    # Asking the user to input the player IDs.
    p1_addl = str(input("Input the first player ID: ")).lower()
    p2_addl = str(input("Input the second player ID: ")).lower()
    
    # Creating a subset where at least one of the players played.
    subset = data[data['Player-additional'].isin([p1_addl, p2_addl])]
    
    # Getting the dates where both players played.
    overlapped_dates = subset[subset.duplicated('Date', keep=False)].sort_values("Date").reset_index()
    
    # Selecting the relevant columns from the data.
    overlapped_dates = overlapped_dates.loc[:,['Player','PTS','TRB','AST','3P','FGA','FG','3PA','3P', 'Player-additional']]
    
    # Creating additional columns for combined stats.
    overlapped_dates['PTS + RB'] = overlapped_dates["PTS"] + overlapped_dates['TRB']
    overlapped_dates['PTS + AST'] = overlapped_dates["PTS"] + overlapped_dates['AST']
    overlapped_dates['RB + AST'] = overlapped_dates["TRB"] + overlapped_dates['AST']
    overlapped_dates['PTS + RB + AST'] = overlapped_dates["PTS"] + overlapped_dates['TRB'] + overlapped_dates["AST"]
    
    # Getting the data for each player.
    player1_data = get_player_data(p1_addl, overlapped_dates)
    player2_data = get_player_data(p2_addl, overlapped_dates)
    
    # Combining the data for both players.
    beef = pd.concat([player1_data,player2_data],axis = 1)
    
    # Renaming the columns to include the player names.
    names = [f"{player1_data.iloc[0,1]} " + i for i in beef.columns][0:14] + [f"{player2_data.iloc[0,1]} " + i for i in beef.columns][14:]
    final = beef.set_axis(names, axis=1)
    
    # Selecting only the numeric columns for the correlation calculation.
    final = final.select_dtypes(include=np.number)
    
     # Calculating the correlation matrix.
    corr_players = final.corr()
    
    # Getting the number of stats for each player.
    num_stats = len(corr_players) // 2
    
    # Selecting only the lower left quadrant of the correlation matrix.
    corr_players = corr_players.iloc[1+num_stats:, 1 :num_stats-1]
    
    # Plotting the correlation matrix.
    plt.figure(figsize=(20, 15))
    sns.heatmap(corr_players,
                annot = True,
                fmt = '.2f',
                cmap='Blues',
                annot_kws={"size": 20})  # Adjust size as needed
    plt.title(f'Correlation between {player1_data.iloc[0,1]} and {player2_data.iloc[0,1]}')
    plt.show()



# Calling the function to start the program.
playerCorrelations()
```
